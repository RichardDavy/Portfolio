## Data preparation

The focus of this project is on the exchagne rate between the Euro and the American Dollar, and so data preparation will solely focus on making the US Dollar data workable.

```python
# We are changing the column titles so that they exclude special characters and are camel case:
exchange_rates.rename(columns={'[US dollar ]': 'US_Dollar', 'Period\\Unit:': 'Time'}, inplace=True)

# We now convert the dates to workable datetimes, using the pandas to_datetime function:
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

# Sorting the values in chronological order:
exchange_rates.sort_values('Time', inplace=True)

# As we have sorted the data, we reset the index values:
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
# Isolating the Time and the US Dollar;
euro_dollar = exchange_rates[['Time','US_Dollar']]

# Checking that we do not have more than one of each date entry (output should be 1):
euro_dollar['Time'].value_counts().max()
```

```python
# Inspecting the US_Dollar column in the same manner as above:
euro_dollar['US_Dollar'].value_counts()
```

We can see that we have 62 entries where the US_Dollar value is  a hypen ('-'). We are going to drop any of these entries from the dataset.

```python
# Keeping the entries that are not hyphens, and repeating the value_counts function to check:
euro_dollar = euro_dollar[euro_dollar['US_Dollar'] != '-']
euro_dollar['US_Dollar'].value_counts()
```

Now that we have removed the hypens, we are going to convert the exchange rate to a float type, which will allow as to look at the numerical properties of the exchange rate:

```python
# Converting the US_Dollar column to float format.
euro_dollar['US_Dollar'] = euro_dollar['US_Dollar'].astype('float')

# Inspecting the converted column.
euro_dollar.info()
euro_dollar.describe()
```

## Eploratory data visualisation.

Now that we have cleaned the data that we want to explore, we can start by producing basic explortory data visuals using __matplotlib.pyplot__.

```python
# Plotting the exchange rate of the Euro against the US Dollar over the total time period:
plt.figure(figsize=(12,6))
plt.plot(euro_dollar['Time'],euro_dollar['US_Dollar'])
plt.title('Euro exchange rate (1999 - 2021)',size=20,weight='bold')
plt.xlabel('Year',size=15)
plt.ylabel('Exchange Rate (USD per Euro)',size=15)
plt.show()
```

We can see that we have plotted the exchange rate for every individual day in the dataset. We may be interested in looking at a rolling average, so that the data represents longer term trends that are not affected by small day-to-day fluctuations.

```python
# Calculating a rolling average over 30 days.
euro_dollar['rolling_mean'] = euro_dollar['US_Dollar'].rolling(30).mean()
```

```python
# Plotting the exchange rate of the Euro against the US Dollar over the total time period with a 30 day rolling average:
plt.figure(figsize=(12,6))
plt.plot(euro_dollar['Time'],euro_dollar['rolling_mean'])
plt.title('Euro Exchange Rate (1999 - 2021) (30 day rolling mean)',size=20,weight='bold')
plt.xlabel('Year',size=15)
plt.ylabel('Exchange Rate (USD per Euro)',size=15)
plt.show()
```

We are going to look at a variety of different rolling means, to figure out what ones are best to investigate different time periods.

```python
# Creating the overall figure.
plt.figure(figsize=(16,12))

# Looping over six different rolling averages to plot. Two variable - integer i and roll which is the rolling mean in days.
for i, roll in zip(range(1,7), (1,7,30,50,100,365)):
    euro_dollar['rolling_mean'] = euro_dollar['US_Dollar'].rolling(roll).mean()
    plt.subplot(3,2,i)
    plt.plot(euro_dollar['Time'],euro_dollar['rolling_mean'])
    plt.title(('{} day rolling average').format(roll), weight='bold', size=14)
    plt.ylabel('USD per Euro')
    sns.despine()
```

From the above, it appears that a rolling mean of 30 days is a good compromise to look at both short and long term trends in the Euro-USD exchange rate.

```python
# Returning the rolling mean over 30 days to the rolling mean column.
euro_dollar['rolling_mean'] = euro_dollar['US_Dollar'].rolling(30).mean()
```

# Explaining the Trends in the USD/EU Exchange Rate

Getting into the geopolitical events that effected the USD/EU exchange rate is beyond the scope of this project. However, the succint summary below was source from [here](https://www.fxempire.com/news/article/the-effect-of-us-presidents-on-eurusd-movement-393715).

## Bill Clinton – The Flourishing Era
“The only way for going wealthier is only by exporting” by words like these Clinton started his ruling, that was during the ratifying of NAFTA agreement in August 1993. In addition to his role in changing the GATT organization and developing it to include more countries with more up to date trading rules.

His policy was courageous and smart, ending up to results the served the American economy. His plan was to cut \\$500B from the budget, \\$225B of them from spending, and raising the rest form taxes including 1.2% special tax imposed on the wealthiest sector and increased the imposed taxed on the self employed sector from 30 to 80%. On the other hand, cutting taxes on small businesses in order to stimulate the economy.  Small business by one time reached 85% of the overall businesses.

By the end of Clinton period, USA had achieved the biggest surplus of governmental spending by 2.5% since the WWI. This was more than normal to drive the Dollar higher, not because of the speculating on the high interest rates, but based on solid land of real national income accompanied by stable inflation rate.

## George W Bush – Numbers Speak for Themselves
We will not delve too much in that era because numbers are more eloquent than the words. Governmental spending increased 70% from \\$1789B to \\$2983B. The Bankruptcy Act 2005, which enforced all home buyers in mortgage to take equity out of their homes. This specific law caused the world financial crisis, followed by stripping out thousands of families their houses. Unemployment rocketed, businesses closed. The EUR/USD saw the highest historical levels by the end of his period.

## Barak Obama – Fixing The Errors
Same as George H Bush, Obama’s policy was focused on getting out of the recession. Unemployment reached 10% by first year and governmental budget reached -10%. In July 2010 the DODD frank law by reforming wall street was approved. Two stimulus plans through taxes cutting first worth \\$787B and the second $858B. The famous TTP(transpacific trade and investment) agreement, and he ended his second period by starting trading partnership negotiations with European allies. By the end of his period, unemployment dropped to 4.7, governmental budget balanced to be -3.2 from -9.8.

EUR/USD pair tumbled during Obama’s two presidential periods, with some kind of long term volatility caused by the global financial crisis.
